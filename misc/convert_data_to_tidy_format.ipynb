```python
import pandas as pd
import numpy as np
```

```python
pubs = ["All", "Atlantic", "Reuters", "New_York_Post", 
                     "Breitbart", "New_York_Times", "Buzzfeed_News",
                     "National_Review", "Business_Insider", "Talking_Points_Memo",
                     "Fox_News", "CNN", "Guardian", "NPR", "Washington_Post"]
df = pd.read_csv("../app/data/Original_Data/highestOccurancesToSentiment{}.csv".format(pubs[0]), sep= ";", index_col= 0)
df.head()
```

```python
# loop
for pub in pubs:
    df = pd.read_csv("../app/data/Original_Data/highestOccurancesToSentiment{}.csv".format(pub), sep= ";", index_col= 0)
    dft = df.T[1:]
    dft["dates"] = dft.index.values
    dft.to_csv("../app/data/Tidy_Data/tidy_sent_data_{}.csv".format(pub))
    
```

```python
# convert occurences
df = pd.read_csv("../app/data/Original_Data/occurances_per_month.csv", sep= ";", index_col= 0)
dft = df.T
dft["dates"] = dft.index.values
dft.to_csv("../app/data/Tidy_Data/tidy_occ_by_time_data.csv")
```
